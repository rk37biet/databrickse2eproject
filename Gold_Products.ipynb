### **DLT Pipeline**

```python
import dlt
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

**Streaming Table**
<br>Create streaming table on a source e.g. silver.products

```python
# Expectations

my_rules = {
    "rule1" : "product_id IS NOT NULL",
    "rule2" : "product_name IS NOT NULL"
}
```

```python
@dlt.table()

@dlt.expect_all_or_drop(my_rules)
def DimProductsStage():
    df = spark.readStream.table("databricks_cata.silver.products")
    return df
```

**Streaming View**

```python
@dlt.view()

def DimProducts_View():
  df = spark.readStream.table("Live.DimProductsStage")
  return df
```

**DimProducts**
<br>
In order to create a dimension we need to use an API called apply_changes

```python
dlt.create_streaming_table("DimProducts")
```

```python
dlt.apply_changes(
    target="DimProducts",
    source="Live.DimProducts_View",
    keys=["product_id"],
    sequence_by="product_id",
    stored_as_scd_type=2
)
```
